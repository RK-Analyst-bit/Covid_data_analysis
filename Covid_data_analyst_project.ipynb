# Active Cases

```python
covid_df
```

```python
 covid_df['Active_cases']=(covid_df['Confirmed']- covid_df['Cured']) -covid_df['Deaths']
```

```python
covid_df
```

```python
covid_df.isnull().sum()
```

```python
covid_df.dropna(inplace = True)
```

```python
covid_df.isnull().sum()
```

```python
covid_df
```

```python
 statewise = pd.pivot_table(covid_df, values = ['Cured','Deaths','Confirmed'], index = 'State/UnionTerritory', aggfunc = max )
```

```python
statewise['Recovery_rate'] = statewise['Cured'] * 100/statewise['Confirmed']
```

```python
statewise['Mortality_rate'] = statewise['Deaths'] * 100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by = 'Confirmed', ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'cubehelix' )
```

# Top 10 states with active cases

```python
covid_df
```

```python
top_10_active_cases=covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by = ['Active_cases'], ascending = False).reset_index()
```

```python
top_10_active_cases
```

```python
plt.figure(figsize = (16,9))
sns.barplot(data =top_10_active_cases.iloc[:10] , x ='State/UnionTerritory' , y= 'Active_cases',linewidth= 2, edgecolor = 'red',palette = "summer")
plt.title('TOP 10 STATES WITH ACTIVE CASE')
plt.xlabel('States')
plt.ylabel('Active Cases')
plt.show()
```

# Top 10  states with most deaths

```python
top_10_states_most_death=covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
```

```python
plt.figure(figsize = (16,9))
sns.barplot(data =top_10_states_most_death.iloc[:10] , x ='State/UnionTerritory' , y= 'Deaths',linewidth= 2, edgecolor = 'black',palette = "inferno")
plt.title('TOP 10 STATES WITH MOST DEATH')
plt.xlabel('States')
plt.ylabel('Deaths')
plt.show()
```

# Growth rate in Active Cases

```python
plt.figure(figsize = (14,9))
sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnatka','Kerala','Uttar pradesh','Tamil Nadu'])],
             x = 'Date',y ='Active_cases', hue = 'State/UnionTerritory')
plt.xlabel('Dates')
plt.ylabel('Active Cases')
plt.title('GROWTH TREND')
plt.show()
```
